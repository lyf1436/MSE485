```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
# ----- change main directory to NPT_Simulations
main_dir = "/Users/kellyh/Documents/UIUC/coursework/mse485-atomistic_scale_simulations/project/MSE485/NPT_Simulations"
TiN_dir = os.path.join(main_dir, "TiN_NPT/data")
ZrN_dir = os.path.join(main_dir, "ZrN_NPT/data")
HfN_dir = os.path.join(main_dir, "HfN_NPT/data")
```

```python
# ------- change the file name -----------
output_dir = os.path.join(ZrN_dir, "data_NoseHoover_1e+6.out")
df = pd.read_fwf(output_dir, delimiter=" ")
df.head()
```

```python
ZrN_df = df[["Step", "Temp", "Lx", "Ly", "Lz", 'PotEng']]

x = ZrN_df["Step"]
temp = ZrN_df["Temp"]
PE = ZrN_df["PotEng"]
lattice = ZrN_df["Lx"]

plt.rcParams.update({'font.size': 15})

plt.plot(x, temp)
plt.xlabel("Step")
plt.ylabel("Temperature (K)")
save_dir = os.path.join(ZrN_dir, "NoseHoover_1e+6_temp.png")
plt.savefig(save_dir, dpi=300, bbox_inches="tight")
plt.show()
print(f"Average Temperature = {np.mean(temp[2000:])}")

plt.plot(x, PE)
plt.xlabel("Step")
plt.ylabel("Potential Energy (eV)")
save_dir = os.path.join(ZrN_dir, "NoseHoover_1e+6_PE.png")
plt.savefig(save_dir, dpi=300, bbox_inches="tight")
plt.show()
print(f"Average Potential Energy = {np.mean(PE[2000:])}")

plt.plot(x, lattice)
plt.xlabel("Step")
plt.ylabel("Lattice constant (A)")
save_dir = os.path.join(ZrN_dir, "NoseHoover_1e+6_lattice.png")
plt.savefig(save_dir, dpi=300, bbox_inches="tight")
plt.show()
print(f"Average lattice constant = {np.mean(lattice[:4000])}")

plt.plot(x, df["Press"])
plt.xlabel("Step")
plt.ylabel("Pressure")
save_dir = os.path.join(ZrN_dir, "NoseHoover_1e+6_pressure.png")
plt.savefig(save_dir, dpi=300, bbox_inches="tight")
plt.show()
```
